```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for Pytorch and Flash
data_pytorch = {
    "Model": ["Small", "Medium", "Large"],
    "Forward Time (s)": [0.001337, 0.004611, 0.02909],
    "Backward Time (s)": [0.002126, 0.006729, 0.08017],
    "Total Time (s)": [0.003463, 0.01134,  0.1093],
    "FLOPS Efficiency (TFLOPs/s)": [4.34, 21.21, 35.22],
    "Peak Memory Usage (MB)": [344.25, 644, 5152]
}
data_flash = {
    "Model": ["Small", "Medium", "Large"],
    "Forward Time (s)": [0.0002036, 0.0046, 0.02853],
    "Backward Time (s)": [0.0006105, 0.006729, 0.08027],
    "Total Time (s)": [0.0008141, 0.0113,  0.1088],
    "FLOPS Efficiency (TFLOPs/s)": [18.47, 21.3, 35.37],
    "Peak Memory Usage (MB)": [80.5, 644, 5152]
}

# Convert data to DataFrames for Pytorch and Flash
df_pytorch = pd.DataFrame(data_pytorch)
df_flash = pd.DataFrame(data_flash)

# Adjusting the bar chart widths for equal spacing

# Define a consistent bar width
bar_width = 0.2

# Plotting Forward, Backward, and Total Times Comparison
plt.figure(figsize=(10, 6))
x = range(len(df_pytorch["Model"]))
plt.bar([i - bar_width / 2 for i in x], df_pytorch["Forward Time (s)"], width=bar_width, color = 'deepskyblue', label="Pytorch Forward Time")
plt.bar([i + bar_width / 2 for i in x], df_flash["Forward Time (s)"], width=bar_width, color = 'tomato', label="Flash Forward Time")
plt.bar([i - bar_width / 2 for i in x], df_pytorch["Backward Time (s)"], width=bar_width, color = 'lightskyblue', label="Pytorch Backward Time", bottom=df_pytorch["Forward Time (s)"])
plt.bar([i + bar_width / 2 for i in x], df_flash["Backward Time (s)"], width=bar_width, color = 'sandybrown', label="Flash Backward Time", bottom=df_flash["Forward Time (s)"])
plt.xticks(x, df_pytorch["Model"])
plt.title("Execution Time Comparison (Pytorch vs Flash)")
plt.xlabel("Model")
plt.ylabel("Time (s)")
plt.legend()
plt.show()

# Plotting FLOPS Efficiency Comparison
plt.figure(figsize=(10, 6))
plt.bar([i - bar_width / 2 for i in x], df_pytorch["FLOPS Efficiency (TFLOPs/s)"], width=bar_width, label="Pytorch FLOPS Efficiency")
plt.bar([i + bar_width / 2 for i in x], df_flash["FLOPS Efficiency (TFLOPs/s)"], width=bar_width, label="Flash FLOPS Efficiency")
plt.xticks(x, df_pytorch["Model"])
plt.title("FLOPS Efficiency Comparison (Pytorch vs Flash)")
plt.xlabel("Model")
plt.ylabel("FLOPS Efficiency (TFLOPs/s)")
plt.legend()
plt.show()

# Plotting Peak Memory Usage Comparison
plt.figure(figsize=(10, 6))
plt.bar([i - bar_width / 2 for i in x], df_pytorch["Peak Memory Usage (MB)"], width=bar_width, label="Pytorch Peak Memory Usage")
plt.bar([i + bar_width / 2 for i in x], df_flash["Peak Memory Usage (MB)"], width=bar_width, label="Flash Peak Memory Usage")
plt.xticks(x, df_pytorch["Model"])
plt.title("Peak Memory Usage Comparison (Pytorch vs Flash)")
plt.xlabel("Model")
plt.ylabel("Memory Usage (MB)")
plt.legend()
plt.show()
```
